## Campo vectorial para dx/dt = sen(x)
Las escalas para x y t son arbitrarias


```python
import numpy as np
from matplotlib import pyplot as plt

nx = 10
nt = 15

x = np.linspace(-np.pi, 2*np.pi, 3*nx+1)
t = np.arange(nt)
```

```python
dt = np.ones([len(t),len(x)]).T
fx = np.tile(np.sin(x), (len(t),1)).T
```

```python
marcas = [r'$-\pi$', r'$-\frac{\pi}{2}$', '0', r'$\frac{\pi}{2}$', r'$\pi$', r'$\frac{3\pi}{2}$', r'$2\pi$']

plt.figure(figsize=(10,8))
plt.quiver(t, x, dt, fx, scale = 25, color = 'b')
plt.xlabel('t', fontsize=20)
plt.ylabel('x', fontsize=20)
plt.title('Campo vectorial para delta_t, delta_f(x)', fontsize=24)
plt.yticks(np.linspace(-np.pi, 2*np.pi, len(marcas)) ,marcas, fontsize=14)
plt.hlines([-np.pi, np.pi], 0, nt, colors = 'g')
plt.hlines([0, 2*np.pi], 0, nt, colors = 'r')
plt.grid()
```
